```python
#scraping data ulasan roblox di playstore
#data yang digunakan berbahasa indonesia
from google_play_scraper import reviews, Sort
app_ID = 'com.roblox.client'

result, _ = reviews(
    app_ID,
    lang='id',
    country='id',
    sort = Sort.NEWEST,
    count= 20000
)

print(f"Jumlah data yang diambil: {len(result)}")
```

```python
import pandas as pd
df_reviews = pd.DataFrame(result)
df_2025 = df_reviews[df_reviews['at'].dt.year == 2025]
df_reviews.head()
```

```python
#menyimpan data reviews dalam csv
df_reviews.to_csv('reviews_roblox.csv', index =False)
```
